```python
from LOBOROBOT import LOBOROBOT  # 载入机器人库

import os

if __name__ == "__main__":
    clbrobot = LOBOROBOT() # 实例化机器人对象
```

```python
    #所有转弯操作可能由于接触面摩擦力影响导致角度出现偏差，可以根据实际需求自行校准
    #小车左拐90度
    #第一个参数是指电机最大转速的30%，可以控制转向速度；第二个参数控制电机运转时间
    clbrobot.turnLeft(30,1.15)  
    clbrobot.t_stop(1) # 机器人停止
```

```python
    #小车右拐90度
    clbrobot.turnRight(30,1.15)  # 机器人右转
    clbrobot.t_stop(1) # 机器人停止
```

```python
    #小车左转右转45度
    clbrobot.turnLeft(30,0.575)  
    clbrobot.t_stop(1) # 机器人停止
    clbrobot.turnRight(30,0.575)  
    clbrobot.t_stop(1) # 机器人停止
```

```python
    #小车前进
    clbrobot.t_up(50,2)      # 机器人前进
    clbrobot.t_stop(1) # 机器人停止
```

```python
    #小车后退
    clbrobot.t_down(50,2)    # 机器人后退
    clbrobot.t_stop(1) # 机器人停止
```
